1.a. Perform language detection


```python
import csv
from langdetect import detect

csv_file_path = r'C:\Users\S SRIHARI\firefox\filtered.csv'


with open(csv_file_path, 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    
    
    for row in reader:
        hindi_text = row[1]

        
        lang = detect(hindi_text)

        
        print("Detected language:", lang)
        print("Hindi text:", hindi_text)
        print("------------------")
```
